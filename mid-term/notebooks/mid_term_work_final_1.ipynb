<h1>let's doing the text preprocessing

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize 
nltk.download('punkt')
```

```python
stop_words = set(stopwords.words('english')) 
len(stop_words)
```

```python
import re
from nltk.stem.porter import  PorterStemmer
ps=PorterStemmer()
```

```python

def remove_stop(x):
  result=[]
  tem1=x.lower()
  tem2=ps.stem(tem1)
  tokens=re.sub(r'[^a-zA-Z]',' ',tem2)
  tem=tokens.split(" ")
  for stop_word in stop_words:
      if stop_word in tem:
          tem.remove(stop_word)
  result.append(" ".join(tem))
  return result
```

```python
x=df['statement'].apply(remove_stop)
# type(x[0])
x[0]
```

```python
x[:10]
```

```python
corpus_1=[]
for tokens in x:
  # main_corups=re.sub(r'\s_',' ',tokens) 
  corpus_1.append(','.join(tokens))
corpus_1
```

```python
# main_corups=re.sub(r'\s_',' ',sentence_) for sentence_ in corpus_ 
# main_corups
```

```python
corpus_=[]
for tokens in corpus_1:
  token=re.sub("\s\s+"," ",tokens)
  if token not in stop_words:
    corpus_.append(''.join(token.strip()))
corpus_
```

<h1><b>Now,create a model

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution1D,Flatten,MaxPool1D
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.layers import Embedding
```

```python
voc_size=10000
onehot_rep=[one_hot(words,voc_size) for words in corpus_ ]
onehot_rep[:10]
```

```python
example_onehot=one_hot('i love baby',voc_size)
example_onehot
```

<h3><b>Embedding Representation

```python
sent_length=30
embedded_docs=pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
embedded_docs
```

```python
embedded_docs.shape
```

<h3>Split the datasets into train and test

```python
from sklearn.model_selection import train_test_split
x_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_final,y_final,test_size=0.33,random_state=42)
```

```python
print(x_train.shape,x_train.shape,y_test.shape,y_train.shape)
print("no.of class:",len(encode_.classes_))
```

```python
embedded_vector_features=40
num_class_=8
# 1st_cnn_kernel =128
# 1st_cnn_size=5
# wnd_size=2

model=Sequential()
model.add(Embedding(voc_size,embedded_vector_features,input_length=sent_length))
# model.add(Dropout(0.5))
model.add(Convolution1D(128,5,padding='valid'))
model.add(MaxPool1D(pool_size=2))
# model.add(Convolution1D())
model.add(LSTM(32))
# model.add(Dropout(0.5))
# model.add(Dense(32,activation='relu'))
# model.add(activation='relu'))
# model.add(Dropout(0.2))
model.add(Dense(num_class_,activation='softmax'))
# model.add(Dense(2))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

<h3>Model Training

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=63,verbose=1)
```

```python
def input_embedded(input_text):
  onehot_rep_input=[one_hot(input_text,voc_size)]
  embedded_docs_input=pad_sequences(onehot_rep_input,padding='pre',maxlen=sent_length)

  input_array=np.array(embedded_docs_input)
  

  return input_array
```
